```python
dtypes = {'ip': 'uint32','app': 'uint16','device': 'uint16','os': 'uint16','channel': 'uint16','is_attributed' : 'uint8'}

def remap_by_freq(feature):
    x = feature.tolist()
    mapping = dict(zip(pd.unique(sorted(x,key=x.count,reverse=True)),range(len(x))))
    feature.replace(mapping, inplace=True)
    return feature  

def add_features(df):
 #This function takes in a dateset and adds the features of diversity of 'app','os','channel' and the click 
 #number for each ip to the dataframe 
 #New features to add: click rate for each ip
 #remapping all features according to frequency
    
 df['app_num'] = df.groupby('ip')['app'].transform('nunique')
 df['device_num'] = df.groupby('ip')['device'].transform('nunique')
 df['os_num'] = df.groupby('ip')['os'].transform('nunique')
 df['channel_num'] = df.groupby('ip')['channel'].transform('nunique')
 df['click_num'] = df.groupby('ip')['ip'].transform('count')
 df['hour'] = pd.to_datetime(df['click_time']).dt.hour.astype('uint8')
 df['app'] = remap_by_freq(df['app'])
 #df['day'] = pd.to_datetime(df['click_time']).dt.day.astype('uint8')

 df.drop(['click_time'], axis=1, inplace=True)
 df.drop(['ip'], axis=1, inplace=True)
 
 return df

#Data loading and processing
train_df = pd.read_csv('../input/train_sample.csv', dtype=dtypes,
                       usecols=['ip','app','device','os', 'channel', 'click_time', 'is_attributed'])

#train_df = add_features(train_df)

print(remap_by_freq(train_df['app']))
#print(train_df['app'])
```
